```python
def sample_language(lang, n=1000):
    ds = load_dataset("ai4bharat/IndicCorpV2", "indiccorp_v2", split=lang, streaming=True)
    return list(itertools.islice(ds, n))

langs = ["hin_Deva", "tel_Telu", "kan_Knda"]
sampled_data = {}

for lang in langs:
    print(f"Sampling {lang} ...")
    sampled_data[lang] = sample_language(lang, n=1000)

print(sampled_data["hin_Deva"][:300])
```
